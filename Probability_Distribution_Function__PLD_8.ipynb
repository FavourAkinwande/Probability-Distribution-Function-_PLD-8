Time Between Arrivals of Buses

Scenario: Buses arrive at a station at an average rate of 3 buses per hour. You want to calculate the probability that the next bus arrives in less than 20 minutes.



```python
#Import neccessary libaries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the rate parameter (lambda)
lambd_rate = 3

# Generate x values (from 0 to 1 hour, with 1000 points)
x_values = np.linspace(0, 1, 1000)

# Exponential PDF: f(x) = lambda * exp(-lambda * x)
pdf = lambd_rate * np.exp(-lambd_rate* x_values)
```

```python
# Plot the Exponential Distribution
plt.plot(x_values, pdf, label=f'Exponential Distribution (Î»={lambd_rate})', color='blue')

# Adding titles and labels
plt.title('Exponential Distribution (PDF) - Time Between Bus Arrivals')
plt.xlabel('Time (hours)')
plt.ylabel('Probability Density')

# Highlight the 20-minute mark (1/3 hour)
plt.axvline(x=1/3, color='red', linestyle='--', label='20 minutes (1/3 hour)')

# Grid and legend for better visualization
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

Scenerio 2 for Python Code : Suppose the time between arrivals of customers at a post office is exponentially distributed with a rate parameter Î»=0.2 customers per minute. This means that, on average, one customer arrives every 5 minutes


```python
import numpy as np

# Define parameters
lambd = 0.2  # rate (customers per minute)
x = 5  # time in minutes between arrivals

# Exponential PDF formula: f(x; Î») = Î» * e^(-Î» * x) using NumPy
pdf = lambd * np.exp(-lambd * x)

# Output the result
print(f"Probability density for a customer arriving exactly after {x} minutes: {pdf:.5f}")
```
